```python
import pandas as pd
```

```python
path = r'D:\viz\d3\static\data\platform_globalsale_all.csv'
path1 = r'D:\viz\d3\static\data\platform_globalsale_new.csv'

df = pd.read_csv(path)
df.head()
```

```python
df1 = df.groupby('Platform').agg(globalsale = ('Global_Sales', sum))
df1['globalsale'] = df1['globalsale'].apply(lambda x: round(x, 2))
df1 = df1.reset_index().rename(columns = {'Platform': 'platform'})
df1 = df1.sort_values(by='globalsale', ascending=False)
df1.to_csv(path1, index=False)

df1
```
